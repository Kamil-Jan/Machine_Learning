```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211) # 2 rows, 1 col, 1st index 
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212) # 2 rows, 1 col, 1st index 
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

```python
x = np.arange(0.1, 30.)
f = np.log(x)**2 - np.sqrt(x)

plt.plot(x, f, 'g-')
plt.yscale('linear') # or e.g. log
# plt.axis([0, 30, -5, 30])
plt.show()
```

```python
data = {i: random.randint(0, (2*(i//2))**2 + i) for i in range(0, 50, 2)}

sizes, prices = zip(*sorted(data.items()))
x = np.arange(0., 50.)
h = 5 + 12*x

plt.plot(sizes, prices, 'rx', x, h)
plt.savefig("plot.png")
plt.show()
```
