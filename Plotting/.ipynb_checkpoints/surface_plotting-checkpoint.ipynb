```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def f(x, z):
    return np.sin(x ** 2 + z ** 2)

x = y = np.arange(-20, 20.5, 1) # creates interval
# meshgrid - returns coordinate matrices from coordinate vectors.
X, Y = np.meshgrid(x, y)

zs = np.array(f(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.jet) # or viridis
ax.set_title('surface')
plt.show()
```

```python
Z = f(X, Y)
plt.contourf(X, Y, Z)
plt.show()
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)
h = plt.contourf(x,y,z)
plt.show()
```
